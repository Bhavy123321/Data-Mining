### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
print("--------------------For PassengerID-------------------")
print("Average -> ", df['PassengerId'].mean())
print("Standard Deviation -> ", df['PassengerId'].std())
print("Maximum -> ", df['PassengerId'].max())
print("Minimum -> ", df['PassengerId'].min())
print("Mode -> ", df['PassengerId'].mode())

print("--------------------Survived-------------------")
print("Average -> ", df['Survived'].mean())
print("Standard Deviation -> ", df['Survived'].std())
print("Maximum -> ", df['Survived'].max())
print("Minimum -> ", df['Survived'].min())
print("Mode -> ", df['Survived'].mode())

print("--------------------Pclass-------------------")
print("Average -> ", df['Pclass'].mean())
print("Standard Deviation -> ", df['Pclass'].std())
print("Maximum -> ", df['Pclass'].max())
print("Minimum -> ", df['Pclass'].min())
print("Mode -> ", df['Pclass'].mode())

print("--------------------Age-------------------")
print("Average -> ", df['Age'].mean())
print("Standard Deviation -> ", df['Age'].std())
print("Maximum -> ", df['Age'].max())
print("Minimum -> ", df['Age'].min())
print("Mode -> ", df['Age'].mode())

print("--------------------SibSp-------------------")
print("Average -> ", df['SibSp'].mean())
print("Standard Deviation -> ", df['SibSp'].std())
print("Maximum -> ", df['SibSp'].max())
print("Minimum -> ", df['SibSp'].min())
print("Mode -> ", df['SibSp'].mode())

print("--------------------Parch-------------------")
print("Average -> ", df['Parch'].mean())
print("Standard Deviation -> ", df['Parch'].std())
print("Maximum -> ", df['Parch'].max())
print("Minimum -> ", df['Parch'].min())
print("Mode -> ", df['Parch'].mode())

print("--------------------Fare-------------------")
print("Average -> ", df['Fare'].mean())
print("Standard Deviation -> ", df['Fare'].std())
print("Maximum -> ", df['Fare'].max())
print("Minimum -> ", df['Fare'].min())
print("Mode -> ", df['Fare'].mode())
```

### 6)  For the qualitative attribute (class), count the frequency for each of its distinct values.



```python
pclass_counts = df['Pclass'].value_counts().sort_index()

print("Frequency of each Passenger Class:")
print(pclass_counts)

Survived = df['Survived'].value_counts().sort_index()

print("Frequency of each Survived Class:")
print(Survived)

Fare  = df['Fare'].value_counts().sort_index()

print("Frequency of each Fare Class:")
print(Fare)

Parch   = df['Parch'].value_counts().sort_index()

print("Frequency of each Parch Class:")
print(Parch)

SibSp    = df['SibSp'].value_counts().sort_index()

print("Frequency of each SibSp Class:")
print(SibSp)

Age    = df['Age'].value_counts().sort_index()

print("Frequency of each Age Class:")
print(Age)
```

###  7) It is also possible to display the summary for all the attributes simultaneously in a table using the describe() function. If an attribute is quantitative, it will display its mean, standard deviation and various quantiles (including minimum, median, and maximum) values. If an attribute is qualitative, it will display its number of unique values and the top (most frequent) values.

```python
print("--------Quantitative-------")
print(df.describe())
print("\n")

print("--------Qualitative-------")
print(df.describe(include="object"))
print("\n")
```

### 8)  For multivariate statistics, you can compute the covariance and correlation between pairs of attributes.



```python
numeric_df = df.select_dtypes(include=['number'])

print("Covariance Matrix:")
print(numeric_df.cov())
print("\n")
```

```python
numeric_df = df.select_dtypes(include=['number'])


print("Correlation Matrix:")
print(numeric_df.corr())
```

### 9) Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
import matplotlib.pyplot as plt

age = df['Age'].dropna()
plt.figure(figsize=(8,5))
plt.hist(age,bins=8,edgecolor='black')
plt.show()
```

### 10) A boxplot can also be used to show the distribution of values for each attribute.

```python
import seaborn as sns

numeric_cols = df.select_dtypes(include='number')
numeric_cols_clean = numeric_cols.dropna()

plt.figure(figsize=(12, 8))
sns.boxplot(data=numeric_cols_clean, orient='h', palette='Set2')
plt.title("Boxplot of Numeric Attributes (Titanic Dataset)")
plt.xlabel("Value")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
pairs = [
    ('Age', 'Fare'),
    ('Pclass', 'Fare'),
    ('Age', 'Pclass'),
    ('SibSp', 'Parch'),
    ('Fare', 'Survived')
]

plt.figure(figsize=(15, 12))
for i, (x, y) in enumerate(pairs, 1):
    plt.subplot(3, 2, i)
    sns.scatterplot(data=df, x=x, y=y, alpha=0.6, color='steelblue', edgecolor='gray')
    plt.title(f'{x} vs {y}')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
