## Now Define Function to  Calculate Entropy

```python
def entropy(y):
    values,counts = np.unique(y,return_counts=True)
    print(values)
    print(counts)
    probabilities = counts / counts.sum()
    print(probabilities)
    return -np.sum(probabilities * np.log2(probabilities))
  
```
